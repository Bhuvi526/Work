```python
class student:

    #default constructors
    def __init__(self):
        pass

    #parameterized constuctors
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        print("adding new student in database..")
```

```python
class student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def get_avg(self):
        sum = 0
        for val in self.marks:
            sum += val
        print("hi", self.name, "your avg score is:", sum/3)
```

```python
s1 = student("bhuvan", [89, 98, 69])
s1.get_avg()
```

```python
s1.name = "gaurav"
s1.get_avg()
```
