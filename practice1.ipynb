```python
numbers = [1,2,3,4,5]
names = ["Bhuvan", "Gaurav", "Nitesh", "Ankush", "Jayesh"]

my_dict = dict(zip(numbers, names))
print(my_dict)
```

```python
def create_dictionary(numbers, names):
    my_dict = {}
    for num, name in zip(numbers, names):
        my_dict[num] = name
    return my_dict

numbers = [1,2,3,4,5]
names = ["Bhuvan", "Gaurav", "Nitesh", "Ankush", "Jayesh"]

result = create_dictionary(numbers, names)
print(result)
```
